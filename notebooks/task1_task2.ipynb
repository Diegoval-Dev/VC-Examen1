# Task 1 – Análisis de Arquitecturas Convolucionales

## Datos del problema

Entrada:
- Dimensiones espaciales: $W = 256$, $H = 256$
- Canales de entrada: $C_{in} = 4$
- Canales de salida deseados: $C_{out} = 64$

Fórmula de salida espacial:

$$
O = \left\lfloor \frac{W - K + 2P}{S} \right\rfloor + 1
$$

---

## 1. Dimensión espacial de salida

### Opción A (Estilo AlexNet)

- Kernel: $7 \times 7$
- Stride: $2$
- Padding: $0$

$$
O = \left\lfloor \frac{256 - 7 + 2(0)}{2} \right\rfloor + 1
$$

$$
O = \left\lfloor \frac{249}{2} \right\rfloor + 1 = 124 + 1 = 125
$$

Salida:

$$
125 \times 125 \times 64
$$

---

### Opción B (Estilo VGG)

Tres convoluciones:
- Kernel: $3 \times 3$
- Stride: $1$
- Padding: $1$

Cada convolución mantiene dimensión espacial:

$$
O = \frac{256 - 3 + 2(1)}{1} + 1 = 256
$$

Después de tres convoluciones:

$$
256 \times 256
$$

MaxPooling:
- Kernel: $2$
- Stride: $2$

$$
O = \frac{256 - 2}{2} + 1 = \frac{254}{2} + 1 = 127 + 1 = 128
$$

Salida final:

$$
128 \times 128 \times 64
$$

---

## 2. Número de parámetros entrenables

Fórmula:

$$
\text{Params} = (K^2 \cdot C_{in}) \cdot C_{out} + C_{out}
$$

(incluye bias)

### Opción A

$$
(7^2 \cdot 4) \cdot 64 + 64
$$

$$
(49 \cdot 4) \cdot 64 + 64
$$

$$
196 \cdot 64 + 64 = 12544 + 64 = 12608
$$

Total:

$$
\boxed{12,608\ \text{parámetros}}
$$

---

### Opción B

Conv1 $(4 \rightarrow 64)$:

$$
(9 \cdot 4) \cdot 64 + 64 = 2368
$$

Conv2 $(64 \rightarrow 64)$:

$$
(9 \cdot 64) \cdot 64 + 64 = 36928
$$

Conv3 $(64 \rightarrow 64)$:

$$
(9 \cdot 64) \cdot 64 + 64 = 36928
$$

Total:

$$
2368 + 36928 + 36928 = 76224
$$

$$

## 3. Justificación teórica

Un kernel $7 \times 7$ tiene campo receptivo de 7.

Tres convoluciones $3 \times 3$ apiladas generan:

$$
RF = 3 + 2 + 2 = 7
$$

Por lo tanto, ambas opciones tienen campo receptivo efectivo equivalente.

Sin embargo, la Opción B introduce:
- Mayor profundidad
- Tres funciones no lineales (ReLU)
- Mayor capacidad de representación
- Mejor abstracción jerárquica

Esto incrementa el poder expresivo del modelo.

Aunque la Opción B tiene más parámetros y mayor costo computacional, ofrece:
- Mejor capacidad de modelar patrones complejos
- Mayor estabilidad en entrenamiento profundo
- Mejor generalización empíricamente demostrada

Por estas razones, es mejor el uso de bloques pequeños y profundos (VGG) en lugar de kernels grandes únicos.

# Task 2 – Criterio Técnico

## 1. Justificación técnica frente a SIFT/HOG + SVM

Los descriptores tradicionales como SIFT y HOG se basan en gradientes locales y estructuras geométricas de bajo nivel. Aunque son robustos a pequeñas transformaciones geométricas, fallan en entornos agrícolas reales donde existen variaciones significativas de iluminación, sombras no uniformes, cambios de ángulo, deformaciones naturales de las hojas y oclusiones parciales causadas por otras hojas o frutos. Estos métodos no modelan semántica; únicamente capturan patrones locales fijos. En presencia de ruido ambiental, texturas irregulares y variabilidad biológica, los descriptores pierden estabilidad y el clasificador SVM depende fuertemente de la calidad manual del feature engineering.

En contraste, una red profunda entrenada mediante Transfer Learning aprende mapas de características jerárquicos. Las primeras capas capturan bordes y texturas; las intermedias detectan patrones estructurales como nervaduras o manchas; y las profundas codifican representaciones semánticas invariantes a iluminación, rotación y variabilidad intra-clase. Estos mapas de características no son descriptores manuales, sino funciones aprendidas que optimizan directamente la separación de clases en un espacio de alta dimensión, permitiendo robustez ante variabilidad real del entorno agrícola.

---

## 2. Elección entre AlexNet y VGG-16

Elegiría VGG-16. Aunque ambas arquitecturas introducen no-linealidad mediante ReLU, VGG-16 posee mayor profundidad y una estructura modular basada en bloques repetitivos de convoluciones $3 \times 3$. Esta modularidad permite incrementar progresivamente la capacidad representacional sin recurrir a kernels grandes, manteniendo estabilidad numérica y mejor control del campo receptivo efectivo.

La mayor profundidad implica más composiciones no lineales, lo que incrementa la capacidad de abstracción jerárquica. Mientras AlexNet utiliza kernels más grandes y menos capas, VGG-16 acumula múltiples transformaciones no lineales que permiten modelar patrones complejos y variaciones semánticas finas en las hojas. En un problema de enfermedades vegetales, donde las diferencias entre clases pueden ser sutiles y dependientes de textura, la mayor capacidad de abstracción profunda de VGG-16 ofrece una representación más discriminativa y robusta.
